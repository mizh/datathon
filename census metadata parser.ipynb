```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import re
import gc

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 2000)

sns.set(context='notebook', style='whitegrid', palette='Set2', font_scale=1.3)
```

```python
def metadata_parser(dataframe):
    
    # grab all the rows in the first columm that start with a capital letter
    headerRows = dataframe[dataframe.iloc[:,0].apply(lambda x: (not x.isnumeric()) and (x[0]==x[0].upper()))]
    headerRows = headerRows.iloc[2::]
    
    summary = dataframe.iloc[[0]] # basic description of the dataset
    
    tables = dict()
    currIndex = 1
    k = 0
    
    # create a dict full of dataframes for each subtable in the metadata file
    for nextIndex in headerRows.index:
        newTable = dataframe.loc[currIndex+1:nextIndex-1,:]
        newTable.columns = dataframe.loc[currIndex,:]
        tables[k] = newTable
        currIndex = nextIndex
        k += 1
        
    return summary, tables
```

```python
files = os.listdir('data')
```

```python
metadataFiles = [file if re.match(r'[0-9]*_MetaData.csv', file) else np.nan for file in files]
metadataFiles = pd.Series(metadataFiles).dropna()
```

```python
metadataFiles.shape[0] # number of census datasets 
```

```python
for file in metadataFiles:
    summary, tables = metadata_parser(pd.read_csv('data/'+file))
    dimTable = tables[0]
    geoDim = dimTable[dimTable['Dimension name']=='Geography']['Dimension ID'].values[0]
    categoryTable = tables[1]
    categoryTable = categoryTable[categoryTable['Dimension ID']!=geoDim] # omit the big list of geographies
    noteTable = tables[6]
    
    display(summary)
    display(dimTable)
    display(categoryTable)
    #display(noteTable) the notes take up a lot of space but they're important to know
```
